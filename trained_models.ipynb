```python
from PyPDF2 import PdfFileReader

import glob
import os

#For reading the CSV's
import csv

#For making a database
import sqlite3
```

```python
DATA_PATH_TEST = os.getcwd()+'/test/'
DATA_PATH_TRAIN = os.getcwd()+'/train/'

def extract_pdf(train):
    for filename in glob.glob(os.path.join(train, '*.pdf')):
        pdf_file = open(filename, 'rb')
        pdf_reader = PdfFileReader(pdf_file)
        text = ''

        for num in range(0, pdf_reader.numPages):
            page = pdf_reader.getPage(num)
            text += page.extractText()
        print("extracted "+str(pdf_reader.numPages)+" pages from "+os.path.basename(filename))
        pdf_file.close()
        yield filename, text
    
    print('done.')

try:
    data_test = {filename.split('\\')[-1].replace('.', ''): text for filename, text in extract_pdf(DATA_PATH_TEST)}
    data_train = {filename.split('\\')[-1].replace('.', ''): text for filename, text in extract_pdf(DATA_PATH_TRAIN)}
    print(data_test.keys())
    print(data_train.keys())
except Exception as e:
    print("An error occurred!")
```

```python
chars_amount_painted_docs = {}

with open('chars_gelakt.txt') as chars_gelakt:
    content = chars_gelakt.readlines()

# remove '/n' from end of string
content = [x.strip() for x in content] 

for line in content:
    #split the items from every line and store them in a list
    document_info = line.split()
    
    #delete index and first percentage by deleting first two items of the list, because they won't be used
    del document_info[0:2]
    
    #make empty var to add every 'word' in list and make one string to get one document name
    document_name = ""
    
    #add all list items to document_name string, except the last one (this is the character amount)
    for word in document_info[:-1]:
        if document_name == "":
            document_name = document_name + word
        else:
            document_name = document_name + " " + word
    
    number_of_characters = document_info[-1]
    
    chars_amount_painted_docs[document_name.replace('.pdf', 'pdf')] = number_of_characters
    
# print dictionary length
print(len(chars_amount_painted_docs))
print("")

# print first five items of dict
counter = 0
for k,v in chars_amount_painted_docs.items():
    if counter < 5:
        print(k, v)
        counter = counter + 1
```

```python
chars_amount_unpainted_docs = {}

with open('chars_ongelakt_1.txt') as chars_ongelakt_1:
    content = chars_ongelakt_1.readlines()
    
with open('chars_ongelakt_2.txt') as chars_ongelakt_2:
    content = content + chars_ongelakt_2.readlines()

# remove '/n' from end of string
content = [x.strip() for x in content] 

for line in content:
    #split the items from every line and store them in a list
    document_info = line.split()
    
    #delete index and first percentage by deleting first two items of the list, because they won't be used
    del document_info[0:2]
    
    #make empty var to add every 'word' in list and make one string to get one document name
    document_name = ""
    
    #add all list items to document_name string, except the last one (this is the character amount)
    for word in document_info[:-1]:
        if document_name == "":
            document_name = document_name + word
        else:
            document_name = document_name + " " + word
    
    number_of_characters = document_info[-1]
    
    chars_amount_unpainted_docs[document_name.replace('.pdf', 'pdf')] = number_of_characters
    
# print dictionary length
print(len(chars_amount_unpainted_docs))
print("")

# print first five items of dict
counter = 0
for k,v in chars_amount_unpainted_docs.items():
    if counter < 5:
        print(k, v)
        counter = counter + 1
```

```python
percentage_painted_per_document = {}

doc_names_painted_docs = list(chars_amount_painted_docs.keys())
doc_names_unpainted_docs = list(chars_amount_unpainted_docs.keys())

for painted_doc_name in doc_names_painted_docs:
    for unpainted_doc_name in doc_names_unpainted_docs:
        if painted_doc_name == unpainted_doc_name:
            try:
                chars_amount_painted_doc = int(chars_amount_painted_docs.get(painted_doc_name))
                chars_amount_unpainted_doc = int(chars_amount_unpainted_docs.get(unpainted_doc_name))

                percentage = 100 - ((chars_amount_painted_doc / chars_amount_unpainted_doc) * 100)
            
                if percentage > 0:
                    percentage_painted_per_document[painted_doc_name] = round(percentage, 2)
            except:
                pass
                
print(len(percentage_painted_per_document))
print(" ")

# print first five items of dict
counter = 0
for k,v in percentage_painted_per_document.items():
    if counter < 5:
        print(k, v)
        counter = counter + 1
```

```python
documenten_train = list(data_train.keys())
documenten_test = list(data_test.keys())
print(documenten_train)
print(documenten_test)
```

```python
x_test = []
for d in documenten_test:
    if d in percentage_painted_per_document:
        x_test.append(data_test[d])
        
y_test = []
for d in documenten_test:
    if d in percentage_painted_per_document:
        y_test.append(percentage_painted_per_document[d])
x_train = []
for d in documenten_train:
    if d in percentage_painted_per_document:
        x_train.append(data_train[d])
        
y_train = []
for d in documenten_train:
    if d in percentage_painted_per_document:
        y_train.append(percentage_painted_per_document[d])
    
print(len(y_test), len(x_test), len(x_train), len(y_train))
print(y_test, y_train)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

cv = TfidfVectorizer(binary=False)#CountVectorizer(binary=False)
# cv_test = CountVectorizer(binary=True, min_df=1, max_df=1)
x_occurences_train = cv.fit_transform(x_train)
# x_occurences_train

x_occurences_test = cv.transform(x_test)

# x_occurences_test
```

```python
inverse_voc = {value: key for key, value in cv.vocabulary_.items()}
print(inverse_voc)
```

```python
col_names = [inverse_voc[i] for i in range(x_occurences_train.shape[1])]
print(col_names)
```

```python
from sklearn.linear_model import Ridge
model = Ridge()

# print(len(x_occurences_train.A))
# print(len(y_train))

# model.fit(x_occurences_train, y_train)


from sklearn.linear_model import Ridge
import matplotlib
import matplotlib.pyplot as plt

# #############################################################################
# Compute paths

n_alphas = 200
# alphas = np.logspace(-10, 20, n_alphas)
alphas = [0.00000000000001, 0.000000001, 0.00000001, 0.0000001, 0.001, 0.1,
         1, 10, 100, 1000, 10000, 100000, 1000000, 10000000, 100000000, 1000000000,
          10000000000, 100000000000, 1000000000000]

coefs = []

for a in alphas:
    ridge = Ridge(alpha=a, fit_intercept=False)
    ridge.fit(x_occurences_train, y_train)
    coefs.append(ridge.coef_)

# #############################################################################
# Display results
 
ax = plt.gca()
ax.plot(alphas, coefs)
ax.set_xscale('log')
plt.xlabel('alpha')
plt.ylabel('coefficient')
plt.title('Ridge Regression')
plt.axis('tight')
plt.show()

```
